<h3>GANs na Geração de Imagens (Gerando Quadros Inéditos do Picasso)</h3>

<h4>Treinamento</h4>

```python
from Neuraline.ArtificialIntelligence.DeepLearning.generative_adversarial_networks import GenerativeAdversarialNetworks
from Neuraline.Utilities.chart import Chart
generative_adversarial_networks, chart = GenerativeAdversarialNetworks(), Chart()

print('[treinando o modelo, aguarde...]')
result = generative_adversarial_networks.addFit(url_path='./picasso')
if result: print('[...modelo treinado com sucesso]')
else: print('[erro no treinamento do modelo.]')
```

<h4>Predição</h4>

```python
print('[gerando imagem, aguarde...]')
result = generative_adversarial_networks.predict(precision=1)
if result:
    chart.plotIMAGE('./predicted.jpg')
    print('[imagem gerada com sucesso.]')
else: print('[erro na geração da imagem.]')
```
